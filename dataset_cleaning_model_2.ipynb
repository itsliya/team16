### bureau

```python
bureau.info()
```

```python
L = [col for col in bureau.columns if bureau[col].dtype == object]

for col in L:
    print(col, 'has ',bureau[col].nunique(), ' distinct values.')
```

Drop all the credit_types that do not go with our target audience 

* 'Loan for working capital replenishment' (entrepreneurs)
* 'Loan for business development' (entrepreneurs)
* 'Real estate loan' (too high of a credit and businessmen)
* 'Interbank credit' (definitely not microcredit)
* 'Loan for purchase of shares (margin lending)' 
* 'Mortgage' (too high of a credit)
* 'Car loan', (too high of a credit)
* 'Unknown type of loan' (can be weird so lets err on the cautious side)

```python
pd.unique(bureau["CREDIT_TYPE"])
```

```python
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Loan for working capital replenishment'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Loan for business development'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Real estate loan'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Interbank credit'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Loan for purchase of shares (margin lending)'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Mortgage'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Car loan'].index, inplace = True)
bureau.drop(bureau[bureau['CREDIT_TYPE'] == 'Unknown type of loan'].index, inplace = True)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(bureau.corr())
```

```python
bureau.drop(["DAYS_ENDDATE_FACT","DAYS_CREDIT_ENDDATE"], axis = 1, inplace = True )
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(bureau.corr())
```
